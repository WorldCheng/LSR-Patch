```python
import numpy as np

# 创建一个实数信号
signal = np.array([1.0, 2.0, 3.0, 4.0])

# 使用 rFFT
result = np.fft.rfft(signal)

print(result)
# 输出长度会是 len(signal)//2 + 1，而不是完整的 len(signal)
```
